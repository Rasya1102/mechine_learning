```python
import pandas as pd
```

```python
data = pd.read_csv('golf_label_num.txt')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
x = data.drop(['Decision'], axis=1)
y = data['Decision']
```

```python
x.shape, y.shape
```

```python
!pip install mlxtend
import joblib
import sys
sys.modules['sklearn.externals.joblib'] = joblib
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
fs = sfs(linear, k_features=3, forward=True, verbose=2, scoring='neg_mean_squared_error')
```

```python
fs = fs.fit(x, y)
```

```python
feat_names = list(fs.k_feature_names_)
print(feat_names)
```

```python
data_baru = data[feat_names]
data_baru['Decission'] = data['Decision']

data_baru.head()
```
